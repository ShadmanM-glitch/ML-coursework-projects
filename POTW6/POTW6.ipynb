#### Feature Columns
    
* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the houseâ€™s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

## Scaling and Train Test Split

TAKE RANDOM STATE =101 AND SPLIT THE DATA SET. TESTING DATA SHOULD BE 30%

```python
X_features = df.iloc[:,1:17]
Y_features = df.iloc[:,0]
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y_features, test_size = 0.3, random_state = 101)
```

```python
print(X_train.shape)
```

### Scaling

IMPORT MINMAXSCALAR AND APPLYING SCALING TO X_TRAIN AND X_TEST

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Creating a Model

IMPORT THE REQUIRED LIBRARIES OR PACKAGES FOR SETTING UP ANN

```python
import tensorflow as tf
```

THERE SHOULD BE 4 HIDDEN LAYERS AND 1 OUTPUT LAYER
COMPILE THE MODEL USING APPROPRIATE MODEL AND LOSS FUNCTION.

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=16, activation='relu')) #selected 16 neurons as I have 16 features in my training data
model.add(tf.keras.layers.Dense(units=16, activation='relu'))
model.add(tf.keras.layers.Dense(units=16, activation='relu'))
model.add(tf.keras.layers.Dense(units=16, activation='relu'))
model.add(tf.keras.layers.Dense(units=1))
```

## Training the Model

TRAIN THE MODDEL. ASSUME BATCH SIZE=128 and EPOCHS SHOULD BE MINIMUM 200


```python
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.fit(X_train, Y_train, batch_size = 128, epochs = 200,validation_data=(X_test,Y_test.values))
```

```python
losses = pd.DataFrame(model.history.history)
```

PLOT LOSSES CALCULATED IN PREVIOUS STEP

```python
sns.lineplot(data=losses,lw=3)
```

# Evaluation on Test Data

https://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
model.evaluate(X_test, Y_test, verbose =0)
```

#### Predicting on Brand New Data

DISPLAY X_TEST
AND MAKE THE PREDICTION
CALCULATE ABSOLUTE ERROR, MEAN SQUARED ERROR, EXPLAINED_VARIANCE_SCORE, MEAN PRICE, MEDIAN OF PRICE

```python
y_pred = model.predict(X_test)
```

```python
print('Absolute Error: ',mean_absolute_error(Y_test,y_pred))
print('Mean Squared Error: ',mean_squared_error(Y_test,y_pred))
print('Root Mean Squared Error: ',np.sqrt(mean_squared_error(Y_test,y_pred)))
print('Explained Variance Score: ',explained_variance_score(Y_test,y_pred))
print('Mean Price: ',np.mean(y_pred))
print('Median Price: ',np.median(y_pred))
```

```python
# CREATE A SCATTERPLOT OF Y_TEST Vs PREDICTIONS
plt.scatter(Y_test,y_pred)

#  predictions
plt.plot(Y_test,Y_test,'r')
plt.xlabel("Predictions")
plt.ylabel("Actual")
plt.title("SCATTERPLOT OF Y_TEST vs PREDICTIONS")
#  predictions
# WHAT DID YOU ANALYZE FROM THE GRAPH
#the model can predict house prices fairly accurately for cheaper houses, but as the prices increases the accuracy decreases
```
