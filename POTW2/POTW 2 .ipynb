# Correlated Feature Check
Helper function that displays correlation by color

```python
def plot_corr(df, size=11):
    """
    Function plots a graphical correlation matrix for each pair of columns in the dataframe.

    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot

    Displays:
        matrix of correlation between columns.
        
        YOU CAN FOLLOW THE FOLLOWING STEPS FOR HINTS
        
    """

    # 1. calculate dataframe correlation  using corr() function
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))# 2. The subplots command in the background will call plt.figure(), and any keywords will be passed along
    ax.matshow(corr)   # 3. USE heatmap with matshow, color code the rectangles by correlation value
    plt.xticks(range(len(corr.columns)), corr.columns)  # 4.  draw x tick marks
    plt.yticks(range(len(corr.index)), corr.index, rotation=40)  # 5. draw y tick marks

plot_corr(df) # 6. Plot the correlation using plot_corr() function.
    

```

# Splitting the Data
* Less training data -> your parameter estimates have greater variance
* Less testing data -> your performance statistic will have greater variance

### Divide data such that neither variance is too high

* Less data -> chances of no satisfactory variance
* More data -> split doesn't really matter

X = feature, independent, predictor
Y = predicted, dependent


Spliting the data¶

70% for training, 30% for testing


Steps required for this problem:
X should have all the columns required for the prediction of the Diabetes.
Y include the column of prediction.
Define the testing percentage as 30%.
By the end of this phase there should be 4 divisions of data: X_train, Y_train, X_test, Y_test


```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state=42)
```

# Impute Missing Values

## Missing Data
1. Ignore
2. Drop rows
3. Replace values (Impute)

```python
#Are there 0 values possible?
#How many rows have have unexpected 0 values?
print('# rows in dataframe', df.shape[0])
glucose_conc = (df['glucose_conc'] == 0).sum()
print('# rows missing glucose_conc:', glucose_conc)
diastolic_bp = (df['diastolic_bp'] == 0).sum()
print('# rows missing diastolic_bp:', diastolic_bp)
thickness = (df['thickness'] == 0).sum()
print('# rows missing thickness:', thickness)
insulin = (df['insulin'] == 0).sum()
print('# rows missing insulin:', insulin)
bmi = (df['bmi'] == 0).sum()
print('# rows missing bmi:', bmi)
diab_pred = (df['diab_pred'] == 0).sum()
print('# rows missing diab_pred:', diab_pred)
age = (df['age'] == 0).sum()
print('# rows missing age:', age)
```

```python
#Impute with the mean
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=0, strategy='mean')
imputer.fit(df.iloc[:,1:10])
df.iloc[:, 1:10] = imputer.transform(df.iloc[:,1:10])
```
